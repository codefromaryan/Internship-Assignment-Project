```python
# MATHEMATICAL METHODS THROUGH NUMPY
arr = np.array([1,2,3,4,5])
print(f'Sum of all elements of array: {np.sum(arr)}')
print(f'Mean of all elements of array: {np.mean(arr)}')
print(f'Median of all elements of array: {np.median(arr)}')
print(f'Standard deviation of all elements of array: {np.std(arr)}')
print(f'Variance of all elements of array: {np.var(arr)}')
print(f'Minimum of all elements of array: {np.min(arr)}')
print(f'Maximum of all elements of array: {np.max(arr)}')
print(f'Product of all elements of array: {np.prod(arr)}')
```

```python
arr = np.array([[10,20,30],
               [40,50,60],
               [70,80,90]])

print(arr[1,1])
print(arr[0,2])
print(arr[2,:3])
print(arr[:, 2])
```
