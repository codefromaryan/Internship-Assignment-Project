###**Agglomerative Clustering**
It is an hierchial clustering where each point starts as its own clusters and then clusters are merged step by step until one big cluster remains.

**Components of AC**
* Data Points: The item theat we want to group.
* Distance metrics: How we measure similarities (Euclidean, Manhattan)
* Linkage method: It decides the distance between the clusters.
    * Single Linkage: Closest pair between the clusters.
    * Complete Linkage: Farthest pair between the clusters.
    * Average Linkage: Average distance between the clusters.
* Dendograms: It is a tree like diagram that shows clusters are getting merged.


**Steps**
1. Start with each data points as its own cluster.
2. We will calculate the distance between all clusters.
3. We will merge the closest two clusters.
4. We will re-calculate the distance between the new cluster that are formed.
5. We are going to repeat it 3-4 times until all the points are in one cluster.
6. We will use dendograms to decide how many clusters to keep.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Clustering Libraries**

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs
```

We will create our dummy data

```python
x,y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=3)
```

```python
x
```

```python
y
```

Creating a dataframe out of the above array

```python
df = pd.DataFrame(x,y)
df.head()
```

```python
df = df.rename(columns={0: 'Feature 1', 1: 'Feature 2'})
```

```python
df.head()
```

Now lets see the relationship between the features

```python
plt.scatter(x[:, 0], x[:, 1], label = y)
plt.xlabel('X: Feature 1')
plt.xlabel('Y: Feature 2')
plt.show()
```

Using dendograms

```python
link = linkage(df, method = 'ward', metric = 'euclidean')
```

```python
plt.figure(figsize = (10, 7))
dendrogram(link)
plt.show()
```

Building the model

```python
model = AgglomerativeClustering(linkage = 'ward', metric = 'euclidean')
```

```python
y_pred = model.fit_predict(df)
```

```python
plt.scatter(df['Feature 1'], df['Feature 2'], c = y_pred, cmap = 'plasma')
plt.show()
```

##**Gaussian Mixter Model (GMM)**
A GMM is a statiscal model which is used to group data into clusters. It assumes that the data comes from multiple gassian distribution ( Normal Distribution) that are combined together.

Think of it as soft clustering technique where each data point can belong to multiple class with their different probabilitic values, unlike K means clustering where a point can belong to only one cluster.

##**Why is GMM used?**
* When clusters have  different shapes and size GMM can be used as it can find elliptical, circular, or any other shape, while K means can only handle spherical cluster.
* When we need soft clustering we can use GMM because it finds the probability of belonging to each clusters.
* Flexible clustering can be seen in GMM because it overlaps the clusters well.


##**How does a GMM Work**
1. Assumes gaussian distribution: It assumes that the data is generated from a mix of gaussian distributed categories

2. Identify the parameters: Each gaussian distribution is characterised by:
    * Mean: Where the center of cluster is
    * Variance/covariance: How spread out the cluster is
    * Weights: How much feature should we consider for gaussian cluster

3. Expectation – Maximization (EM):
    * Expectation step: We will calculate the probability that each point belongs to a specific gaussian cluster (soft assignment ot clusters)
    * Maximization step: Now we will adjust the parameters (mean, variance, weight) of gaussian to better fit the data

4. Repeat: We have to repeat it until the parameters got stablizied or the clusters stop changing significantly.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

We will create the synthetic data

```python
np.random.seed(42)
x,y_true = make_blobs(n_samples=500, centers=3, cluster_std=[1.0,2.5,0.5], random_state=42)
```

```python
x
```

```python
y_true
```

Building GMM model

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmmLabels = gmm.fit_predict(x)
```

```python
gmmLabels
```

Silhoutte score

```python
silhoutteGMM = silhouette_score(x, gmmLabels)
silhoutteGMM
```

Lets see with K means

```python
km = KMeans(n_clusters=3, random_state=42)
kmLabels = km.fit_predict(x)
```

```python
kmLabels
```

```python
silhoutteKM = silhouette_score(x, kmLabels)
silhoutteKM
```

```python
fig, axes = plt.subplots(1,3, figsize = (20,5), sharex = True, sharey = True)
axes[0].scatter(x[:,0], x[:,1], c=y_true, cmap='viridis', s = 30)
axes[0].set_title('True Clusters')
axes[1].scatter(x[:,0], x[:,1], c=gmmLabels, cmap='viridis', s = 30)
axes[1].set_title('GMM Clusters')
axes[2].scatter(x[:,0], x[:,1], c = kmLabels, cmap='viridis', s = 30)
axes[2].set_title('KM Clusters')
plt.show()
```
