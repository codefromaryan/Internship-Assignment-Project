Graphs

```python
import re
```

```python
sentence = "My no. is 12345-67897 and email is something@gmail.com"

numberPattern = r"\d{5}-\d{5}"

numMatch = re.search(numberPattern, sentence)
```

```python
if numMatch:
  print('Phone:',numMatch.group())
```

```python
re.findall(numberPattern, sentence)
```

```python
emailPattern = r"\S+@\S+\.\S+"
emailMatch = re.search(emailPattern, sentence)
```

```python
if emailMatch:
  print('Email:',emailMatch.group())
```

```python
re.findall(emailPattern, sentence)
```

```python
import matplotlib.pyplot as plt
```

1) Line Plot

```python
month = ['Jan', 'Feb', 'March', 'Apr', 'May']
savings = [8000, 9000, 12000, 7000, 10000]
savings2 = [12000, 2300, 3222, 6000, 2355]

plt.plot(month, savings, savings2, marker = '.')
plt.show()
```

2) Bar Graph


```python
month = ['Jan', 'Feb', 'March', 'Apr', 'May']
savings = [8000, 9000, 12000, 7000, 10000]

plt.bar(month, savings, )
plt.show()
```

```python
month = ['Jan', 'Feb', 'March', 'Apr', 'May']
savings = [8000, 9000, 12000, 7000, 10000]

plt.barh(month, savings,color = 'red')
```

3) Histogram

```python
marks = [10, 40, 30, 55, 23, 67, 40, 60]
plt.hist(marks, bins = 8, edgecolor = 'black')
plt.show()
```

4) Scatter Plot


```python
h = [150, 160,165,170,180,195]
w = [50,55,60,65,70,80]
plt.scatter(h, w)
plt.show()
```

5) Pie Chart


```python
from enum import auto
month = ['Jan', 'Feb', 'March', 'Apr', 'May']
savings = [8000, 9000, 12000, 7000, 10000]

plt.pie(savings, labels = month, autopct="%0.2f%%")
plt.show()
```

Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
sns.lineplot(data=df,x='size', y = 'total_bill')
plt.show()
```

```python
sns.barplot(data=df,x='day', y = 'tip', ci = False)
plt.show()
```

```python
sns.countplot(data=df, x='day')
plt.show()
```

```python
sns.histplot(data=df, x='total_bill', bins = 10, kde = True)
plt.show()
```

6)Boxplot


```python
sns.violinplot(data=df, y='tip')
plt.show()
```

```python
sns.scatterplot(data=df, y='total_bill', x='tip')
plt.show()
```

Pairplot


```python
sns.pairplot(df)
plt.show()
```
